# Apprentissage par l'ACTION — actif vs passif

Un agent **pousse** des disques (masses cachées) et apprend leur physique depuis les **pixels bruts** (world model = CNN + ensemble de prédicteurs, auto-supervisé, zéro objet codé en dur).

- **Actif** : il pousse là où ses prédicteurs sont en **désaccord** (curiosité).
- **Passif** : il pousse au **hasard**.

**L'edge testé** : à interactions égales, la courbe **active** doit apprendre la physique **plus vite** que la passive. Aucun téléchargement. Runtime **GPU**.
